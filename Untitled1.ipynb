```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from scipy.signal import find_peaks
```

```python
def generateSpectrum(Nbins=1000, Npeaks=5, noise=0.05):
    energy = np.linspace(0, 10, Nbins)
    spectrum = np.zeros(Nbins)

    for _ in range(Npeaks):
        peak_center = np.random.uniform(1, 9)
        peak_height = np.random.uniform(10, 100)
        peak_width = np.random.uniform(0.1, 0.5)
        spectrum += peak_height * np.exp(-0.5 * ((energy - peak_center) / peak_width) ** 2)

    noise = np.random.normal(0, noise * spectrum.max(), size=Nbins)
    spectrum += noise
    spectrum = np.clip(spectrum, 0, None)

    return energy, spectrum
```

```python
Nspectra = 50  # Number of synthetic spectra
energyBins, _ = generateSpectrum()  # Energy bins remain constant
allSpectra = []
for i in range(Nspectra):
    _, spectrum = generateSpectrum()
    allSpectra.append(spectrum)
allSpectra = np.array(allSpectra)
```

```python
def getFeatures(spectrum):
    peaks, _ = find_peaks(spectrum, height=np.mean(spectrum) + np.std(spectrum))
    num_peaks = len(peaks)
    mean_intensity = np.mean(spectrum)
    variance = np.var(spectrum)
    skewness = np.mean((spectrum - mean_intensity)**3) / (np.std(spectrum)**3 + 1e-8)
    return [num_peaks, mean_intensity, variance, skewness]
```

```python
features = []
for spectrum in allSpectra:
    features.append(getFeatures(spectrum))

pca = PCA(n_components=2)
reducedFeatures = pca.fit_transform(features)

silhouetteScores = []
Krange = range(2, 8)

for k in Krange:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(reducedFeatures)
    score = silhouette_score(reducedFeatures, labels)
    silhouetteScores.append(score)
```

```python
plt.figure(figsize=(8, 6))
plt.plot(Krange, silhouetteScores, marker='o')
plt.title("Silhouette Scores for K")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Silhouette Score")
plt.show()
```

```python
optimal_k = Krange[np.argmax(silhouetteScores)]
print(f"Optimal Number of Clusters (K): {optimal_k}")
```

```python
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
clusters = kmeans.fit_predict(reducedFeatures)
```

```python
plt.figure(figsize=(8, 6))
colors = cm.get_cmap('tab10', optimal_k)
for cluster in range(optimal_k):
    plt.scatter(reducedFeatures[clusters == cluster, 0],
                reducedFeatures[clusters == cluster, 1],
                label=f"Cluster {cluster+1}", c=colors(cluster))
plt.title("K-Means Clustering on Synthetic X-ray Spectroscopy Data")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
for cluster in range(optimal_k):
    # Get the indices of spectra belonging to the current cluster
    indices = np.where(clusters == cluster)[0]
    clusterIndx1 = indices[1]
    clusterIndx2 = indices[2]

    plt.plot(energyBins, allSpectra[clusterIndx1], label=f"Cluster {cluster+1}", c= colors(cluster))
    plt.plot(energyBins, allSpectra[clusterIndx2], label=f"Cluster {cluster+1}", c= colors(cluster), alpha=0.5)

plt.title("Sample Spectra from Each Cluster")
plt.xlabel("Energy (keV)")
plt.ylabel("Intensity (Counts)")
plt.legend()
plt.show()
```
